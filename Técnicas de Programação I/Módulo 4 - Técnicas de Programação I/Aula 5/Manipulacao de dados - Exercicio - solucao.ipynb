## 2. Cruze essas bases usando como chave a coluna "id_casa" para que tenhamos um único DataFrame para trabalhar.

```python
df_merged = pd.merge(df_casas_final, df_email, on='id_casa', how='inner')
df_merged
```

```python
df_merged.drop_duplicates(inplace=True)
```

## 3. A média de preço varia de acordo com a quantidade de banheiros? Mostre.

```python
df_merged.groupby(by=['FullBath'])['SalePrice'].mean()
```

## 4. Transforme a coluna "SalePrice" em uma coluna categórica, com as seguintes categorias: ['cheap', 'medium', 'expensive']. OBS: Crie uma nova coluna, não substitua a antiga.

```python
## Definindo meus intervalos
bins = pd.IntervalIndex.from_tuples([(0, 100000), (100000, 300000), (300000, 1000000)])
bins
```

```python
df_merged['SalePrice_cut'] = pd.cut(df_merged['SalePrice'], bins=bins)
df_merged.head()
```
