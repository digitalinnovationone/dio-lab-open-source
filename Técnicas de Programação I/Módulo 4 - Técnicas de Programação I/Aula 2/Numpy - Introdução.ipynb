# Numpy

Quando falamos em computao cientfica, uma das bibliotecas mais importantes  o Numpy. Basicamente, Numpy  uma biblioteca python que permite a criao do que conhecemos como *ndarrays*, que so vetores que permitem realizar operaes com alta perfomance e, em geral, sem loops explcitos. Devido a sua velocidade  muito utilizada no mundo de Data Science.

## Mas por qual motivo os *ndarrays* so mais rpidos que as listas em python?

Para isso, vamos entender como a lista armazena seus elementos:

**Lista**



**ndarray**



Ou seja, no *ndarray* salvamos diretamente o elemento e no uma referncia dele. Dessa forma, sabendo o tipo de *ndarray* que estamos usando (inteiros, float, etc) sabemos exatamente quanto de espao precisamos e alm disso no precisamos ficar verificando o tipo do elemento, como na lista, onde sabemos apenas a referncia do elemento e precisamos verificar de que tipo ele , o que  um procedimento mais custoso que no numpy.

Alm disso, o Numpy  escrito na linguagem C, que  uma linguagem bem mais performtica.

Esses fatores em conjunto fazem com que ele seja uma opo extremamente rpida para efetuar operaes vetoriais e matriciais.

## 6. Operaes bsicas

O Numpy nos permite fazer operaes entre arrays e entre array e escalar! Tudo isso de um jeito muito direto e sem necessitar de loops.

````python
array_1 = np.array([0, 1, 2, 3, 4, 5])
array_1 * 3
````

````python
array_1
````

````python
array_1 = array_1 * 3
array_1
````

````python
array_1 = np.array([0, 1, 2, 3, 4, 5])
array_1 + 4
````

````python
array_1 - 6
````

Essas operaes entre arrays e escalares o numpy chama de *Broadcasting*.



````python
# Operacoes entre array

array_1 = np.array([0, 1, 2, 3, 4, 5])
array_2 = np.array([5, 4, 3, 2, 1, 0])

array_1 + array_2
````

````python
array_1 * array_2
````

````python
array_3 = np.full(6, 20)
array_3
````

````python
array_1 / array_3
````
