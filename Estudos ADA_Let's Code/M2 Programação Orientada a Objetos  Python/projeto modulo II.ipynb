```python
class Registro:
    '''
    Cria objeto de registro de ponto do funcionário
    argumentos: h = hora
                m = minuto
    '''
    # Define os atributos do objeto da classe Registro
    def __init__(self, hora, minuto):
        self.h = hora
        self.m = minuto
   # Define representação da hora no estilo 00:00     
    def __repr__(self):
        
        return f"{self.h:02d}:{self.m:02d}"
        
   # processo de subtração de duas classes Registro     
    def __sub__(self, other):
        ho = self.h - other.h
        mi = self.m - other.m
        if mi >= 60:
            ho = ho + 1
            mi = mi - 60
        return Registro(ho, mi)
    
        
```
