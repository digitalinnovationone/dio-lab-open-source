```python

def fizzBuzz(n):
    
    
    
    for i in range(1, n+1):
    
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
    
        elif i % 3 == 0 and not i % 5 == 0:
            print("Fizz")

        elif i % 5 == 0 and not i % 3 == 0:
            print("Buzz")

        else:
            print(str(i))
        
    return
    
```

```python
fizzBuzz(15)
```

```python
def fizzbuzz(n):

    if n % 3 == 0 and n % 5 == 0:
        return 'FizzBuzz'
    elif n % 3 == 0:
        return 'Fizz'
    elif n % 5 == 0:
        return 'Buzz'
    else:
        return str(n)

print "\n".join(fizzbuzz(n) for n in xrange(1, 21))
```
