# Hadoop em arquiteturas modernas

Criar uma pasta na raiz do DBFS chamada Aula

```
dbutils.fs.mkdirs('/aula')
```

Listar o conteúdo do diretório recém criado

```
dbutils.fs.ls('/aula')
```

d
 Criar arquivo txt chamado hadoop no diretório /aula usando comando put com um conteúdo qualquer

```
dbutils.fs.put('/aula/hadoop.txt', 'Hands on no Databricks')
```

Listar o conteúdo do diretório /aula

```
dbutils.fs.ls('/aula')
```

Exibir os 10 primeiros bytes do arquivo de texto recém criado

```
dbutils.fs.head('/aula/hadoop.txt',10)
```

Criar um diretório chamado aula_hadoop dentro de /aula

```
dbutils.fs.mkdirs('/aula/aula_hadoop')
```

Mover o arquivo aula/hadoop.txt para aula/aula_hadoop

```
dbutils.fs.mv('/aula/hadoop.txt','/aula/aula_hadoop')
```

Copiar o arquivo /aula/aula_hadoop/hadoop.txt para /aula

```
dbutils.fs.cp('/aula/aula_hadoop/hadoop.txt','/aula')
```

Remover o diretório /aula recursivamente

```
dbutils.fs.rm('/aula', recurse = True)
```
