## Tabebela Gerenciada

Crie um banco de dados chamado aula_db

```
%sql
CREATE DATABASE aula_db;
```

Qual foi o resultado no DBFS?

Um diretório foi criado em: /user/hive/warehouse chamado aula_db.db.

Crie uma tabela chamada "aluno" no banco de dados aula_db

```
%sql
CREATE TABLE aula_db.aluno (id INT, nome STRING);
```

Qual foi o resultado no DBFS?

Um novo diretório foi criado em: /user/hive/warehouse/aula_db.db chamado aluno.

Insira um primeiro registro na tabela recém criada

```
%sql
INSERT INTO aula_db.aluno VALUES (1,'Matheus');
```

Qual foi o resultado no DBFS?

Um arquivo foi criado e uma pasta foram criados em: /user/hive/warehouse/aula_db.db/aluno

Insira um segundo registro na tabela recém criada

```
%sql
INSERT INTO aula_db.aluno VALUES (2, 'Bowie')
```

Qual foi o resultado no DBFS? E qual o formato dos arquivos?

Um segundo arquivo foi criado em: /user/hive/warehouse/aula_db.db
