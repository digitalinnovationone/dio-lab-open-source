## Exercício 1:
Utilizando a base de dados da IBM e a biblioteca plotly:

**a)** Faça um gráfico de dispersão (scatter plot) de tamanho (10,5), com a variável *YearsAtCompany* no eixo X e a variável *DailyRate* no eixo Y. Coloque o título "YearsAtCompany x DailyRate" na cor azul e com uma letra diferente da usual.

```python
## Link para ajudar a mexer na cor e fonte do título: https://plotly.com/python/figure-labels/

fig = px.scatter(data_frame=df1, x='YearsAtCompany', y='DailyRate', width= 960, height=480, ##tamanho em pixels
                title='YearsAtCompany x DailyRate')
fig.update_layout(title=dict(font=dict(family="cursive", color='blue')))
fig.show()
```

**b)** Repita o gráfico do item a, identificando as observações do gênero feminino com a cor vermelha e do gênero masculino com a cor verde. Não esqueça de inserir legenda para as cores.

```python
## Link para ajudar a mexer na cor e fonte do título: https://plotly.com/python/figure-labels/

fig = px.scatter(data_frame=df1, x='YearsAtCompany', y='DailyRate', width= 960, height=480, ##tamanho em pixels
                title='YearsAtCompany x DailyRate',
                color='Gender', color_discrete_map={'Female' : 'red', 'Male':'green'})
fig.update_layout(title=dict(font=dict(family="cursive", color='blue')))
fig.show()
```

**c)** Repita o gráfico do item a, porém agora identificando os funcionários pela variável 'JobSatisfaction', 
utilizando a seguinte escala de cores:
- cor: '#FFE4E1' para 'Low' 
- cor: '#FFB6C1' para 'Medium' 
- cor: '#FF69B4' para 'High' 
- cor: '#C71585' para 'Very High' 

```python
## Link para ajudar a mexer na cor e fonte do título: https://plotly.com/python/figure-labels/

df_aux = df1.copy()
df_aux['JobSatisfaction'] = df_aux['JobSatisfaction'].map({1:'Low',
                                                          2:'Medium',
                                                          3:'High',
                                                          4:'Very High'})

fig = px.scatter(data_frame=df_aux, x='YearsAtCompany', y='DailyRate', width= 960, height=480, ##tamanho em pixels
                title='YearsAtCompany x DailyRate',
                color='JobSatisfaction', 
                 color_discrete_map={'Low' :'#FFE4E1', 'Medium':'#FFB6C1', 'High':'#FF69B4', 'Very High':"#C71585"})
fig.update_layout(title=dict(font=dict(family="cursive", color='blue')))
fig.show()
```

## Exercício 2

**a)** Utilizando a base de dados de covid e a biblioteca plotly, crie um gráfico de linhas mostrando a evolução do número de casos novos mês a mês, para as DRS 'Grande São Paulo', 'São José do Rio Preto' e 'Marília'. Identifique os DRS por cores e legenda. Não esqueça de colocar título para o gráfico e para os eixos. 

```python
df_aux = df2[df2['nome_drs'].isin([ 'Grande São Paulo', 'São José do Rio Preto', 
                                   'Marília'])].groupby(['nome_drs', 'ano_mes'], as_index=False).sum()

fig = px.line(data_frame=df_aux, x='ano_mes', y='casos_novos', color='nome_drs', title='Casos novos por mês e DRS')
fig.update_layout(xaxis_title="<b>Mês</b>",
    yaxis_title='<b>Casos Novos</b>')
fig.show()
```

**b)** Repita o gráfico do item a trocando as cores dos DRS.

```python
df_aux = df2[df2['nome_drs'].isin([ 'Grande São Paulo', 'São José do Rio Preto', 
                                   'Marília'])].groupby(['nome_drs', 'ano_mes'], as_index=False).sum()

fig = px.line(data_frame=df_aux, x='ano_mes', y='casos_novos', color='nome_drs', 
              title='Casos novos por mês e DRS', color_discrete_map={'Grande São Paulo' : 'red',
                                                                    'Marília' : 'purple',
                                                                    'São José do Rio Preto' : 'green'})
fig.update_layout(xaxis_title="<b>Mês</b>",
    yaxis_title='<b>Casos Novos</b>')
fig.show()
```

**c)** Considerando apenas os DRS 'Bauru', 'São José do Rio Preto' e 'Marília', faça três gráficos de pizza, para os meses de abril, maio e junho, mostrando a porcentagem de novos óbitos em cada DRS.

```python
df_aux = df2[df2['nome_drs'].isin(['Bauru', 'São José do Rio Preto','Marília']) &
            df2['ano_mes'].isin(['2020-04','2020-05','2020-06'])].groupby(['nome_drs','ano_mes'], as_index=False).sum()

fig = px.pie(data_frame=df_aux[df_aux['ano_mes']=='2020-04'], names='nome_drs', values='obitos_novos',
            title='Óbitos novos por DRS em abril/20',  width= 500, height=480)
fig.show()

fig = px.pie(data_frame=df_aux[df_aux['ano_mes']=='2020-05'], names='nome_drs', values='obitos_novos',
            title='Óbitos novos por DRS em maio/20',  width= 500, height=480)
fig.show()

fig = px.pie(data_frame=df_aux[df_aux['ano_mes']=='2020-06'], names='nome_drs', values='obitos_novos',
            title='Óbitos novos por DRS em junho/20',  width= 500, height=480)
fig.show()
```

## Exercício 3:
Utilizando a base de dados de covid e a biblioteca plotly:

**a)** Faça um gráfico de barras mostrando o total de óbitos (considerando todos os meses) para todos os DRS, excluindo a Grande São Paulo. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom e intercale as barras com as cores 'red' e 'green'.

```python
df_aux = df2[df2['nome_drs'] != 'Grande São Paulo'].groupby('nome_drs', as_index=False).sum()
df_aux['color'] = ['1','2']*8
```

```python
Auto-Indent Lines (Ctrl + Alt + I) and Reformat Code (Ctrl + Alt + L) 
```

```python
fig = px.bar(data_frame=df_aux, x='nome_drs', y='obitos_novos', color='color', color_discrete_map={'1':'red',
                                                                                                  '2':'green'},
            category_orders={'nome_drs': [i for i in df_aux['nome_drs']]})
fig.update_xaxes(tickangle=-45)
fig.update_layout(showlegend=False)
fig.show()
```

**b)**  Faça um gráfico de barras horizontal mostrando o total de óbitos no mês de abril para 3 DRS da sua escolha. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom.

```python
df_aux = df2[df2['nome_drs'].isin(['Araraquara', 'Bauru', 'Piracicaba'])
   & (df2.ano_mes=='2020-04')].groupby(['nome_drs'], as_index=False).sum()

fig = px.bar(data_frame=df_aux, x='obitos_novos', y='nome_drs', orientation='h', title='Óbitos por DRS no mês de abril/20')
fig.show()
```
