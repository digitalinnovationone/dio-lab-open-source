O Seaborn suporta vários temas que podem alterar o estilo dos gráficos muito fácil e economizar muito tempo. Usando a função set_style() do Seaborn podemos definir qualquer um dos temas disponíveis na biblioteca Seaborn. Aqui estão alguns dos temas populares:

* Darkgrid
* Whitegrid
* Dark
* White
* Ticks

```python
# alterando o estilo em um plot do matplotlib 
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'])
sns.set_style('whitegrid')
plt.show()
```

Vamos melhorar um pouco esse visual:

```python
# alterando o estilo e removendo eixos 
# alterando o estilo em um plot do matplotlib 
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'])
sns.set_style('ticks')
sns.despine()
plt.show()
```

Um outro parâmetro interessante é o contexto, por default, ele é setado como notebook, mas existem alguns outros: 

* Paper
* Notebook
* Talk
* Poster

```python
# alterando o contexto 
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'])
sns.set_style('ticks')
sns.set_context('poster')
sns.despine()
plt.show()
```

##  Vamos ao SNS Scatter

Relação entre duas variáveis continuas

```python
# criando uma visualizacao com o scatterplot
sns.set_style("dark")
sns.set_context("notebook")
sns.scatterplot(data=df, x= 'TotalWorkingYears', y = 'MonthlyIncome');
# plt.show()
```

Usando o set, podemos alterar nomes dos labels e colocar título:

```python
# adicionando labels e titulos
g = sns.scatterplot(data=df, x= 'TotalWorkingYears', y = 'MonthlyIncome')
g.set(xlabel= 'Anos Trabalhados', ylabel = 'Salario', title= 'Comparacao', );

```

Mas também podemos usar o matplotlib

```python
# adicionando labels e titulos com o matplotlib 
import matplotlib.pyplot as plt
sns.scatterplot(data=df, x= 'TotalWorkingYears', y = 'MonthlyIncome')
plt.title('Comparacao')
plt.xlabel('Anos trabalhados');
```

Como visto acima, um gráfico de dispersão descreve a relação entre dois fatores. Podemos descrever ainda a relação entre várias variáveis, ou seja, como a variação em uma variável de dados afeta a representação das outras variáveis de dados em um gráfico inteiro.


## Algumas propriedades Propriedades 

### HUE


O parâmetro hue pode ser usado para agrupar as multiplas variáveis de dados e mostrar a dependência entre elas em termos de cores diferentes dos marcadores usados para plotar os valores dos dados

## Distribuições 

Um gráfico de estimativa de densidade do kernel (KDE) é um método para visualizar a distribuição de observações em um conjunto de dados, análogo a um histograma. O KDE representa os dados usando uma curva de densidade de probabilidade contínua em uma ou mais dimensões.

Ele representa a distribuição de probabilidade dos valores de dados como a área sob a curva traçada.

```python
# KDE
sns.kdeplot(data=df, x='Age')
```

```python
# len(df[(df.Age > 30) & (df.Age < 40)])/len(df)
```

```python
df.MonthlyIncome
```

```python
# kde com shade
sns.kdeplot(data=df, x='Age', shade=True,hue='Attrition')
```

```python
# df.select_dtypes(include=np.number)
```

```python
# outra exemplo de kde
# kde com shade
sns.kdeplot(data=df, x='MonthlyIncome')
```

Análises estatísticas requerem conhecimento sobre a distribuição de variáveis em seu conjunto de dados. A função seaborn displot() suporta várias abordagens para visualizar distribuições. Isso inclui técnicas clássicas como histogramas e abordagens computacionalmente intensivas, como estimativa de densidade do kernel:

```python
# displot com col e kde
sns.displot(data=df, x="Age", col="Attrition", kde=True)
```

Seaborn também tenta promover técnicas poderosas, mas menos familiares, como calcular e traçar a função de distribuição cumulativa empírica dos dados:

```python
# distribuicao acumulada ecdf, col e hue 
sns.displot(data=df, x="Age",kind='ecdf')
```

## Plots para dados categóricos

Ao contrário do barplot em que usamos uma variável categórica e uma continua, no catplot count, usamos apenas uma categórica

```python
# catplot com bar
sns.catplot(data=df, kind="bar", x="BusinessTravel", y="Age", hue="Attrition")
```

```python
# catplot count
sns.catplot(data=df, kind="count", x="BusinessTravel")
```

```python
# catplot count horizontal
sns.catplot(data=df, kind="count", y="BusinessTravel")
```

```python
df.describe()
```

```python
# configurando order no catplot count
df.BusinessTravel.unique()
category_order = ['Non-Travel','Travel_Rarely', 'Travel_Frequently']
sns.catplot(data=df, kind="count", x="BusinessTravel", order=category_order)
```

```python
# catplot box 
sns.catplot(x='Attrition', y='MonthlyIncome', data=df, kind='box');
```

```python
# boxplot removendo outliers
sns.catplot(x='Attrition', y='MonthlyIncome', data=df, kind='box', sym='');
```

```python
# alterando o limite dos bigodes 
# boxplot removendo outliers
sns.catplot(x='Attrition', y='MonthlyIncome', data=df, kind='box', whis= [5,95]);
```

```python
# colocando os bigodes entre 0 e 100
sns.catplot(x='Attrition', y='MonthlyIncome', data=df, kind='box', whis= [0,75]);
```

* Podemos querer dar um foco maior na distribuição dos dados ao invés dos outliers

```python
# visualizando a distribuicao no boxplot com boxen
sns.catplot(x="Attrition", y="MonthlyIncome", kind="boxen", data=df);
```

Podemos combinar dois plots e visualizar os pontos dentro do boxplot: 

```python
# duas visualizacoes catplot e stripplot
sns.catplot(x="Attrition", y="MonthlyIncome", kind="boxen", data=df);
sns.stripplot(x="Attrition", y="MonthlyIncome", data=df, color="blue");
```

Aqui também podemos dividir os graficos usando o atributo col: 

```python
# catplot e col no box plot
sns.catplot(x="Attrition", y="MonthlyIncome", kind="box", col = 'Gender', data=df);
```

```python
# distribuicao dos pontos 
sns.catplot(x="BusinessTravel", y="Age", palette="Set2",hue="Attrition", data=df);
```
