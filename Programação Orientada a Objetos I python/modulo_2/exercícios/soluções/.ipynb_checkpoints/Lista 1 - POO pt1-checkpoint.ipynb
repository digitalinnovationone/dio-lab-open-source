1. Crie uma classe `Bola` cujos atributos são cor e raio. 

Crie um método para calcular a área dessa bola (obs.: a área de uma esfera é $4 \pi r^2$). 

Crie um método que imprime a cor da bola. 

Crie um método para calcular o volume da bola (obs.: o volume de uma esfera é $\frac{4}{3}\pi r^3$). 

Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida.

````python
pi = 3.14
class Bola:
    
    def __init__(self, raio, cor):
        self.raio = raio
        self.cor = cor
        
    def area(self):
        return 4*pi*self.raio**2
    
    def volume(self):
        return (4/3)*pi*self.raio**3

# Usando a classe bola:
bolinha = Bola(3,'vermelho')

print("A área é:", bolinha.area())
print("O volume é:", bolinha.volume())
````

2. Crie uma classe `Retangulo` cujos atributos são `lado_a` e `lado_b`. 

Crie um método para calcular a área desse retângulo. 

Crie um objeto dessa classe e calcule a área e a imprima em seguida.


````python
class Retangulo:
    
    def __init__(self, lado_a, lado_b):
        
        self.lado_a = lado_a
        self.lado_b = lado_b
        
    def area(self):
        
        return self.lado_a*self.lado_b

# Exemplo de uso:
retangulo = Retangulo(4,3)

print("A área do retângulo é:", retangulo.area())
````

4. Com base no exercício anterior, crie um sistema de cadastro com dicionários e a classe `Cliente`; cada cliente deve ter como chave o seu CPF. Seu programa deve perguntar se o usuário quer cadastrar um novo cliente, alterar um cadastro ou sair. 

Obs.: tente fazer esse exercício pode criando uma classe `Sistema`, que irá controlar o sistema de cadastros. Essa classe deve ter o atributo cadastro e os métodos para imprimir os cadastrados, cadastrar um novo, alterar um cadastro ou sair.

````python
class Cliente:
    
    def __init__(self, nome, idade, email, cpf):
        
        self.nome = nome
        self.idade = idade
        self.email = email
        self.cpf = cpf
        
    def visualizar(self):
    
        print("Nome:", self.nome)
        print("\nIdade:", self.idade)
        print("\nE-mail:", self.email)
        print("\nCPF:", self.cpf)
        
class SistemaCadastro:
    
    def __init__(self):
        
        self.cadastrados = {}
        
    def cadastrar_novo(self):
        
        nome = input("Digite o nome do cliente: ")
        idade = int(input("Digite a idade do cliente: "))
        email = input("Digite o email do cliente: ")
        cpf = input("Digite o CPF: ")
        novo_cliente = Cliente(nome, idade, email, cpf)
        self.cadastrados[cpf] = novo_cliente
        
    def visualizar_cadastros(self):
        
        for cpf_cliente in self.cadastrados:
            self.cadastrados[cpf_cliente].visualizar()
            print('---------------')
            
    def alterar_cadastro(self, cpf):
        
        cliente = self.cadastrados.pop(cpf)
        alteracao = input('O que deseja alterar?')
        
        while alteracao not in ('nome', 'idade', 'email','cpf'):
            
             input('Erro. Digite novamente o que deseja alterar?')
                
        if alteracao == 'nome':
            
            novo_nome = input("Digite o novo nome: ")
            cliente.nome = novo_nome
            
        if alteracao == 'idade':
            
            nova_idade = int(input("Digite a nova idade: "))
            cliente.idade = nova_idade
            
        if alteracao == 'email':
            
            novo_email = input("Digite o novo e-mail: ")
            cliente.email = novo_email
            
        if alteracao == 'cpf':
            
            novo_cpf = input("Digite o novo CPF: ")
            cliente.cpf = novo_cpf
            
        self.cadastrados[cliente.cpf] = cliente
    
    def run(self):
                   
        o_que_fazer = int(input("O que deseja fazer? "))
        
        while o_que_fazer != 0:
            
            if o_que_fazer == 1:
                
                print("===Novo Cadastro===\n")
                self.cadastrar_novo()
                print('\n===Fim do cadastro===\n')
                
            elif o_que_fazer == 2:
                
                print('===Visualização dos Cadastrados===\n')
                self.visualizar_cadastros()
                print('\n===Fim da visualização===\n')
                
            elif o_que_fazer == 3:
                
                print('===Alteração de Cadastro===\n') 
                cpf_a_alterar = input("Digite o CPF a ser alterado: ")
                
                while cpf_a_alterar not in self.cadastrados:
                    
                    cpf_a_alterar = input("CPF não encontrado. Digite novamente: ")
                    
                self.alterar_cadastro(cpf_a_alterar)
                
                print('\n===Fim da alteração===\n')
                
            o_que_fazer = int(input("O que deseja fazer? "))
            
        print('===Fim===')    

# Exemplo de uso:
sistema = SistemaCadastro()
sistema.run()
````

5. Crie uma classe `Funcionario` cujos atributos são nome e e-mail. 

Guarde as horas trabalhadas em um dicionário cujas chaves são o mês em questão e, em outro dicionário, guarde o salário por hora relativo ao mês em questão. 

Crie um método que retorna o salário mensal do funcionário.


````python
class Funcionario:
    
    def __init__(self, nome, email):
        
        self.nome = nome
        self.email = email
        self.horas = {}
        self.salario_hora = {}
        
    def calcula_salario(self,mes):
        
        if (mes not in self.horas) or (mes not in self.salario_hora):
            raise KeyError('Mês não encontrado.')
            
        return self.horas[mes]*self.salario_hora[mes]

## Exemplo de uso:
funcionario = Funcionario('José da Silva', 'josesilva@letscode.com')
funcionario.horas['10/2019'] = 160 # horas trabalhadas em 10/2019
funcionario.salario_hora['10/2019'] = 25.00 # valor do salário-hora
print(funcionario.calcula_salario('10/2019'))
````

6. Crie uma classe `ContaCorrente` com os atributos cliente (que deve ser um objeto da classe Cliente) e saldo. 

Crie métodos para depósito, saque e transferência. Os métodos de saque e transferência devem verificar se é possível realizar a transação.

````python
class Cliente:
    
    def __init__(self, nome, idade, email, cpf):
        
        self.nome = nome
        self.idade = idade
        self.email = email
        self.cpf = cpf
        
    def visualizar(self):
    
        print("Nome:", self.nome)
        print("\nIdade:", self.idade)
        print("\nE-mail:", self.email)
        print("\nCPF:", self.cpf)
        
        
class ContaCorrente:
    
    def __init__(self, cliente, saldo = 0):
        
        self.cliente = cliente
        self.saldo = saldo
        
        
    def saque(self, valor):
        
        if self.saldo - valor >= 0:
            
            self.saldo -= valor
        else:
            
            print('Saldo insuficiente.')
            
    def deposito(self, valor):
        
        self.saldo += valor
        
        
    def transferencia(self, other, valor):
        
        if self.saldo - valor >= 0:
            
            self.saldo -= valor
            other.saldo += valor
        else:
            
            print('Saldo insuficiente.')
            
        
### exemplo de uso:
##### criando os clientes:
jose = Cliente('José da Silva', 28,'a@a.com','9999999')
maria = Cliente('Maria de Souza', 80, 'b@b.com', '888888')

conta_jose = ContaCorrente(jose, 1000)
conta_maria = ContaCorrente(maria, 25000)

print("Início:")
print("Saldo de José é:",conta_jose.saldo)
print("Saldo de Maria é:",conta_maria.saldo)

##### Sacando 

conta_jose.saque(100)
print("\nApós saque:")
print("Saldo de José é:",conta_jose.saldo)
print("Saldo de Maria é:",conta_maria.saldo)

##### Depositando

conta_maria.deposito(1500)
print("\nApós depósito:")
print("Saldo de José é:",conta_jose.saldo)
print("Saldo de Maria é:",conta_maria.saldo)

##### Transferindo

conta_jose.transferencia(conta_maria, 120)
print("\nApós transferência:")
print("Saldo de José é:",conta_jose.saldo)
print("Saldo de Maria é:",conta_maria.saldo)
````

7. Crie uma classe `Televisor`cujos atributos são: fabricante; modelo; canal atual; lista de canais; volume.

Faça métodos aumentar/diminuir volume, trocar o canal e sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV. 

Obs.: o volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

````python
class Televisor:
    
    def __init__(self, fab, modelo):
        
        self.fabricante = fab
        self.modelo = modelo
        self.canal_atual = None
        self.lista_de_canais = []
        self.volume = 20

    def aumentaVolume(self, valor):
        
        if self.volume + valor <= 100:
            self.volume += valor
        else:
            self.volume = 100

    def diminuiVolume(self, valor):
        
        if self.volume - valor >= 0:
            self.volume -= valor
        else:
            self.volume = 0
    
    def trocaCanal(self, canal):
        
        if canal in self.lista_de_canais:
            self.canal_atual = canal

    def sintonizaCanal(self, canal):
        
        if canal not in self.lista_de_canais:
            self.lista_de_canais.append(canal)
````

9. O módulo (biblioteca) `time` possui a função `time.sleep(x)`, que faz seu programa “dormir” por x segundos. Utilizando essa função, crie uma classe `Cronômetro` e faça um programa que cronometre o tempo.


````python
import time

class Cronometro:
    
    def __init__(self, segundos = 0, minutos = 0, horas = 0):
        
        self.segundos = segundos
        self.minutos = minutos
        self.horas = horas
        
    def __repr__(self):
        
        texto = "%.2d:%.2d:%.2d" %(self.horas, self.minutos, self.segundos)
        return texto
    
    def incremento(self):
        
        if self.segundos >= 60:
            self.segundos = 0
            self.minutos += 1
        
        if self.minutos >= 60:
            self.minutos = 0
            self.horas += 1
        
        self.segundos += 1
        
    def run(self):
        
        ## ele vai ficar em loop infinito!!! tem que parar o processo quando quiser
        while True:
            print('\n') # "limpar a shell". Não funciona no Jupyter!
            print(self)
            self.incremento()
            time.sleep(1)

# Exemplo de Uso:

cronometro1 = Cronometro()
cronometro1.run()
````

10. Crie uma modelagem orientada a objetos para uma agenda capaz de armazenar contatos. Através dessa agenda é possível incluir, remover, buscar e listar contatos já cadastrados. 

Dica: siga o modelo do exercício 4.


````python
class Contato:
    
    def __init__(self, nome, telefone, email, endereco):
        self.nome = nome
        self.telefone = telefone
        self.email = email
        self.endereco = endereco
        
    def visualizar(self):
        
        print("Nome:", self.nome)
        print("\nTelefone:", self.telefone)
        print("\nE-mail:", self.email)
        print("\nEndereco:", self.endereco)
        
        
class Agenda:
    
    def __init__(self):
        
        self.contatos = {}
        
    def cadastrar_novo(self):
        
        nome = input("Digite o nome: ")
        telefone = int(input("Digite o telefone (apenas números): "))
        email = input("Digite o e-mail: ")
        endereco = input("Digite o endereço: ")
        novo_contato = Contato(nome, telefone, email, endereco)
        self.contatos[email] = novo_contato
        
    def visualizar_cadastros(self):
        
        for email in self.contatos:
            
            self.contatos[email].visualizar()
            print('---------------')
            
    def alterar_cadastro(self, email):
        
        contato = self.contatos.pop(email)
        alteracao = input('O que deseja alterar? (nome, telefone, email ou endereco)')
        
        while alteracao not in ('nome', 'telefone', 'email','endereco'):
            
             input('Erro. Digite novamente o que deseja alterar?')
        
        if alteracao == 'nome':
            
            novo_nome = input("Digite o novo nome: ")
            contato.nome = novo_nome
        
        if alteracao == 'telefone':
            
            novo_telefone = int(input("Digite o novo telefone: "))
            contato.telefone = novo_telefone
        
        if alteracao == 'email':
            
            novo_email = input("Digite o novo e-mail: ")
            contato.email = novo_email
        
        if alteracao == 'endereco':
            
            novo_endereco = input("Digite o novo endereço: ")
            contato.endereco = novo_endereco
            
        self.contatos[contato.email] = contato
        
    def run(self):
        
        o_que_fazer = 1
        
        while o_que_fazer != 0:
            
            if o_que_fazer == 1:
                
                print("===Novo Contato===\n")
                self.cadastrar_novo()
                print('\n===Fim do cadastro===\n')
                
            elif o_que_fazer == 2:
                
                print('===Visualização dos Contatos===\n')
                self.visualizar_cadastros()
                print('\n===Fim da visualização===\n')
                
            elif o_que_fazer == 3:
                
                print('===Alteração de Contato===\n') 
                email_a_alterar = input("Digite o e-mail do contato a ser alterado: ")
                
                while email_a_alterar not in self.contatos:
                    
                    email_a_alterar = input("E-mail não encontrado. Digite novamente: ")
                    
                self.alterar_cadastro(email_a_alterar)
                print('\n===Fim da alteração===\n')
                
            o_que_fazer = int(input("O que deseja fazer?"
                                    "\n1 - Cadastrar Novo Contato"
                                    "\n2 - Visualizar Contatos "
                                    "\n3 - Alteração de Contato"
                                    "\n0 - Sair\n"))
            
        print('===Fim===') 

# # Exemplo de Uso:

agenda = Agenda()
agenda.run()
````
