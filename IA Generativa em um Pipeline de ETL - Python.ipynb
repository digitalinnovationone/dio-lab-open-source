# LOAD

```python
import json

def load_users_locally(users, filename="users_updated.json"):
    with open(filename, "w", encoding="utf-8") as file:
        json.dump(users, file, ensure_ascii=False, indent=2)
    return True
```

```python
success = load_users_locally(users)
print(f"Dados carregados localmente? {success}!")
```
