# Hierarchical Clustering:



Existem dois métodos de realizar o agrupamento hierárquico:

1. **Aglomerativa (AGNES - Agglomerative Nesting):** Esta é uma abordagem "de baixo para cima": cada observação começa como seu próprio cluster e os pares de clusters são mesclados de acordo com sua similaridade até que resulte em um único grande cluster.

2. **Divisiva (DIANA - Divise Analysis):** Esta é uma abordagem "de cima para baixo": todas as observações começam em um único cluster e as divisões são executadas recursivamente à medida que se desce na hierarquia. O processo é iterado até que todos os objetos estejam em seu próprio cluster.

<br>

**Quando a clusterização hierárquica é utilizada?** <br>
A clusterização hierárquica é geralmente usada para agrupar uma população em diferentes grupos. Alguns exemplos comuns incluem segmentar clientes, agrupar documentos semelhantes, agrupar animais entre diferentes subgrupos/classes, recomendar músicas ou filmes semelhantes, etc.

## Clustering Hierárquico Divisivo

O agrupamento hierárquico divisivo funciona de maneira oposta. Em vez de começar com n clusters (no caso de n observações), começamos com um único cluster e atribuímos todos os pontos a esse cluster.
Portanto, não importa se temos 10 ou 1000 pontos de dados. Todos esses pontos pertencerão ao mesmo cluster no início.

Agora, a cada iteração, dividimos o ponto mais distante do cluster e repetimos esse processo até que cada cluster contenha apenas um único ponto:</p>

**Etapas para realizar o agrupamento hierárquico**

Agora a questão é – como decidimos quais pontos são semelhantes e quais não são? Essa é uma das questões mais importantes no agrupamento!

No agrupamento hierárquico, temos um conceito chamado `matriz de proximidade`. Isso armazena as distâncias entre cada ponto. Vamos dar um exemplo para entender essa matriz, bem como as etapas para realizar o agrupamento hierárquico.

<br>
Lembrando que as métricas de distância podem ser de diferentes maneiras como já vimos (ex:euclidiana, manhattan, etc.)
 




 

## Vamos a um exemplo!
Suponha que uma professora queira dividir seus alunos em grupos diferentes. Ela tem as notas obtidas por cada aluno em uma tarefa e, com base nessas notas, deseja segmentá-los em grupos. Não há um alvo fixo aqui quanto a quantos grupos ter. Assim, tentaremos aplicar aqui o agrupamento hierárquico e segmentar os alunos em diferentes grupos.

Vamos pegar uma amostra de 5 alunos:

|`Aluno`|`Nota`|
|-----|------|
|`1`|10|
|`2`|7|
|`3`|28|
|`4`|20|
|`5`|35|

<center><img src="https://s3-sa-east-1.amazonaws.com/lcpi/f246dc18-c74d-41e2-8299-dd93d87a5a27.png" width=800></center>

Agora precisamos definir uma métrica de proximidade dos alunos, que no nosso caso será a distância **Euclidiana**. Então, digamos que queremos calcular a distância entre as notas do aluno 1 e 2:

$$\sqrt{(10-7)^2} = \sqrt{9} = 3$$

Da mesma forma, podemos calcular todas as distâncias e preencher a __matriz de proximidade__:

|`ID`|`1`|`2`|`3`|`4`|`5`|
|-----|---|--|---|---|---|
|`1`|0|3|18|10|25|
|`2`|3|0|21|13|28|
|`3`|18|21|0|8|7|
|`4`|10|13|8|0|15|
|`5`|25|28|7|15|0|


Os elementos diagonais desta matriz serão sempre 0, pois a distância de um ponto consigo mesmo é sempre 0.

### Passo-a-passo
__Etapa 1:__ Primeiro, atribuímos todos os pontos a um cluster individual:

__Etapa 2:__ Em seguida, veremos a menor distância na matriz de proximidade (alunos 1 e 2) e mesclaremos os pontos com a menor distância. Em seguida, atualizamos a matriz de proximidade:


<center><img src="https://s3-sa-east-1.amazonaws.com/lcpi/4359a1ab-33b5-4459-b1db-6517f6e8812a.png" width=800></center>

Chegamos na conclusão de unir esses dois clusters. Mas qual será a nota desse novo cluster? Nossa escolha influenciará no método de clusterização

Aqui, tomamos o máximo das duas notas (7, 10) para substituir as notas deste cluster. Em vez do máximo, também podemos obter o valor mínimo, médios ou o centróide.

|`Aluno`|`Nota`|
|-----|------|
|`(1, 2)`|10|
|`3`|28|
|`4`|20|
|`5`|35|

Agora, vamos calcular novamente a matriz de proximidade para esses clusters:

|`ID`|`(1, 2)`|`3`|`4`|`5`|
|-----|---|--|---|---|
|`(1, 2)`|0|18|10|25|
|`3`|18|0|8|7|
|`4`|10|8|0|15|
|`5`|25|7|15|0|

__Etapa 3:__ Vamos repetir a etapa 2 até que reste apenas um único cluster.

Então, vamos primeiro olhar para a distância mínima na matriz de proximidade e então mesclar o par de clusters mais próximo. Obteremos os clusters mesclados conforme mostrado abaixo após repetir essas etapas:


<center><img src="https://s3-sa-east-1.amazonaws.com/lcpi/47136842-0a78-4380-9b56-9f31462a1311.png" width=800></center>


Começamos com 5 clusters e finalmente temos um único cluster. É assim que funciona o agrupamento hierárquico aglomerativo.

Outros conceitos importantes que aprendemos na formação de clusters:
- O que determina a 'clusterização' é a distância (que pode ser calculada de diferentes formas)
- O que determina a formação de novos clusters é o método de clusterização (vimos o valor máximo, mas veremos outros mais comuns em breve)

É assim que podemos decidir o número de clusters usando um dendrograma no Hierarchical Clustering. Além disso, com o dendograma conseguimos saber a ordem em que os clusters foram criados. No nosso caso, o primeiro cluster formado é o (1, 2) seguido por (3,5), (1,2,4) e finalmente (1, 2, 3, 4, 5).

```python
# dataset: https://archive.ics.uci.edu/ml/datasets/wholesale+customers
# forma interessante de extrair o csv: diretamente da internet
df_customers = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
df_customers.head()
```

```python
sns.pairplot(df_customers)
```

Existem várias categorias de produtos – Frescos, Leite, Mercearia, etc. Os valores representam o número de unidades compradas por cada cliente para cada produto. Nosso objetivo é fazer clusters a partir desses dados que possam segmentar clientes semelhantes juntos. Obviamente, usaremos o agrupamento hierárquico para esse problema.

<br>
<br>

Mas antes de aplicar o Hierarchical Clustering, temos que fazer algo importante! O que é?

<br> 
**Escalar** os dados para que a escala de cada variável seja a mesma. Por que isso é importante? Bem, se a escala das variáveis ​​não for a mesma, o modelo pode se tornar tendencioso para as variáveis ​​de maior magnitude, como Fresco ou Leite (consulte a tabela acima).

Então, vamos primeiro normalizar os dados e trazer todas as variáveis ​​para a mesma escala:

```python
from sklearn.preprocessing import MinMaxScaler

df_customers_scaled = MinMaxScaler().fit_transform(df_customers) # escalo os dados
df_customers_scaled = pd.DataFrame(df_customers_scaled, columns=df_customers.columns) # crio dataframe dos dados escalados
df_customers_scaled.head()
```

```python
# documentação dendograma no scipy https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(20, 14))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_customers_scaled, method='complete', metric='euclidean'))

```

O eixo x contém as amostras e o eixo y representa a distância entre essas amostras. A linha vertical com distância máxima é a linha azul e, portanto, podemos decidir um limite de 10 e cortar o dendrograma:

```python
plt.figure(figsize=(20, 14))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_customers_scaled, method='complete', metric='euclidean'))
plt.axhline(y=2, color='r', linestyle='--')
```

Temos dois clusters, pois esta linha corta o dendrograma em dois pontos. Vamos agora aplicar clustering hierárquico para 2 clusters:

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')  
cluster.fit_predict(df_customers_scaled) # lembrando da boa prática: utilizar predict não faz tanto sentido semântico dado que estamos falando de um dataset não supervisionado
```

```python
# faz mais sentido utilizarmo o labels_

cluster.labels_
```

```python
# é só uma questão de boa prática. no final ambos são iguais
np.unique(cluster.labels_ == cluster.fit_predict(df_customers_scaled), return_counts=True)
```

Podemos ver os valores de 0s e 1s na saída (definimos 2 clusters). 0 representa os pontos que pertencem ao primeiro cluster e 1 representa os pontos do segundo cluster. Vamos agora visualizar os dois clusters:

```python
plt.figure(figsize=(10, 7))  
plt.scatter(df_customers_scaled['Milk'], df_customers_scaled['Grocery'], c=cluster.labels_) 
```

```python
data_scaled_with_labels = df_customers_scaled
data_scaled_with_labels['label'] = cluster.fit_predict(df_customers_scaled)
data_scaled_with_labels
```

```python
sns.pairplot(data_scaled_with_labels, hue='label')
```

## Quais são outros métodos de medir similaridade entre dois clusters?

No exemplo das notas dos alunos selecionamos a maior nota como a representante do cluster, mas como medir a similaridade entre dois clusters?

Vários métodos de aglomeração de clusters foram desenvolvidos para responder a essa pergunta. Os mais comuns são:

- __Agrupamento de ligação máximo ou completo (Complete or Maximum Linkage):__ Ele calcula todas as distâncias entre os elementos do cluster 1 e os elementos do cluster 2 e considera o maior valor como a distância entre os dois clusters. Tende a produzir clusters mais compactos.
- __Agrupamento de ligação mínimo ou único (Single or Minimum Linkage):__ calcula todas as diferenças entre os elementos do grupo 1 e os elementos do grupo 2, e considera a menor dessas diferenças como critério de ligação. Ele tende a produzir aglomerados longos e “soltos”.
- __Agrupamento de ligação médio (Average Linkage):__ Ele calcula todas as distâncias entre os elementos do cluster 1 e os elementos do cluster 2, e considera a média delas como a distância entre os dois clusters.
- __Agrupamento de ligação centroide (Centroid linkage):__ calcula a distância entre o centróide para o cluster 1 (um vetor médio de variáveis ​​de comprimento p) e o centróide para o cluster 2.
- __Método de variação mínima de Ward (Ward's Method):__ Minimiza a variação total dentro do cluster. Em cada etapa, o par de clusters com distância mínima entre clusters é mesclado. Que é baseado no aumento do erro quadrado quando dois clusters são mesclados, e é semelhante à média do grupo se a distância entre os pontos for a distância ao quadrado.

Vamos pegar mais um exemplo. Agora dessa vez com uma dataset que já conhecemos, o dataset Iris 🌻

O que podemos falar desse gráfico:

```python
species = iris.frame['target'].replace(iris.target.unique(), iris.target_names).values
species
```

# Conclusões


O agrupamento hierárquico (também chamado de hierarchical cluster analysis or HCA) é:
- uma abordagem alternativa ao agrupamento k-means/DBScan para identificar grupos no conjunto de dados.
- não exige que pré-especifiquemos o número de clusters a serem gerados, conforme exigido pela abordagem k-means.
- o agrupamento hierárquico tem uma vantagem adicional sobre o agrupamento K-means, pois resulta em uma representação atraente baseada em árvore das observações, chamada dendrograma. 
- O processo de agrupamento hierárquico constrói uma ordem entre os clusters.



<br>



Pontos fortes do agrupamento hierárquico

- É fácil de entender e implementar.
- Não precisamos pré-especificar nenhum número específico de clusters.
    - Pode obter qualquer número desejado de clusters cortando o dendrograma no nível adequado.
- Eles podem corresponder a uma classificação significativa.
- Fácil de decidir o número de clusters apenas olhando para o Dendrograma.

<br>

<br>

Limitações do cluster hierárquico

- O clustering hierárquico não funciona bem em grandes quantidades de dados.
- Todas as abordagens para calcular a similaridade entre clusters têm suas próprias desvantagens.
- No clustering hierárquico, uma vez tomada a decisão de combinar dois clusters, ela não pode ser desfeita.
- Diferentes medidas têm problemas com um ou mais dos seguintes.
    - Sensibilidade a ruídos e outliers.
    - Enfrenta Dificuldade ao lidar com diferentes tamanhos de clusters.
    - Está quebrando grandes aglomerados.
    - Nesta técnica, a ordem dos dados tem impacto nos resultados finais.
