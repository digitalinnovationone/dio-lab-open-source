```python
import requests, uuid

# 1. COLE A NOVA CHAVE QUE VOCÊ COPIOU AQUI
chave = "COLE_SUA_KEY_AQUI"
endpoint = "https://api.cognitive.Microsofttranslator.com"
regiao = "brazilsouth" # Ajustado conforme seu print

# 2. CONFIGURAÇÃO
path = '/translate?api-version=3.0&to=en'
headers = {
    'Ocp-Apim-Subscription-Key': chave,
    'Ocp-Apim-Subscription-Region': regiao,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}

body = [{ 'text': 'Eu consegui configurar o tradutor corretamente!' }]

# 3. EXECUÇÃO
request = requests.post(endpoint + path, headers=headers, json=body)
response = request.json()

print("✅ Resultado da Tradução:")
print(response[0]['translations'][0]['text'])
```
