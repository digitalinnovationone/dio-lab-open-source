```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('Aulas/imgs/placa.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'
```

```python
options = "--psm 11"
text = pytesseract.image_to_string(img, config=options)
text
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(3,15),0)
ret1,th1 = cv2.threshold(blur,10,240,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(th1,cmap='gray')
```

```python
options = "--psm 11"
text1 = pytesseract.image_to_string(th1, config=options)
text1
```
